```python
pd.set_option('display.max_colwidth',10000)
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv('test.csv')
df = pd.concat([train_df,test_df])
print(df.describe())
print(df.info())
columns = df.columns
```

```python
df = df.drop(["Alley","FireplaceQu","PoolQC","MiscFeature","Fence"],axis=1)
```

```python
#所有枚举型属性
tdf = train_df
tdf = tdf.drop(["Alley","FireplaceQu","PoolQC","MiscFeature","Fence"],axis=1)
list_a = [item for item in tdf.columns if tdf[item].dtype=='object']
tdf.groupby(tdf.BldgType)['SalePrice'].mean().plot()
for item in list_a:
    #这条命令用于创建新图，不然都会画在一起
    plt.figure()
    tdf.groupby(tdf[item])['SalePrice'].mean().plot()
```
